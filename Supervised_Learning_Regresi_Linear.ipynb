```
# --- SUPERVISED LEARNING: REGRESI LINEAR SEDERHANA ---
# Tujuan: Memprediksi Nilai Ujian berdasarkan Jam Belajar

# 1. MENGIMPOR LIBRARY YANG DIBUTUHKAN
# Library adalah kumpulan kode yang sudah jadi dan bisa kita gunakan ulang.

import numpy as np  # Numpy digunakan untuk operasi angka, terutama untuk membuat array (kumpulan data)
import matplotlib.pyplot as plt  # Matplotlib digunakan untuk membuat grafik atau plot visualisasi data
from sklearn.linear_model import LinearRegression # Scikit-learn (sklearn) adalah library utama untuk machine learning di Python. Kita mengimpor model Regresi Linear.

# 2. MENYIAPKAN DATASET
# Kita membuat data sederhana sendiri untuk latihan.
# Bayangkan ini adalah data dari beberapa siswa.

# X adalah fitur (variabel independen), yaitu data yang kita gunakan untuk memprediksi.
# Di sini, X adalah 'Jam Belajar'.
# np.array([...]) mengubah list biasa menjadi array Numpy.
# .reshape(-1, 1) mengubah format data agar bisa dibaca oleh model scikit-learn.
X = np.array([1, 2, 3, 4, 5, 6, 8, 10]).reshape(-1, 1)

# y adalah label (variabel dependen), yaitu data yang ingin kita prediksi.
# Di sini, y adalah 'Nilai Ujian' yang didapat.
y = np.array([55, 60, 62, 75, 78, 85, 90, 98])

# 3. MEMBUAT DAN MELATIH MODEL
# Di sinilah proses 'belajar' oleh mesin terjadi.

# Membuat objek model Regresi Linear. Anggap saja kita menyiapkan 'otak' kosong.
model = LinearRegression()

# Melatih model dengan data kita. Fungsi .fit() akan mencari pola atau hubungan matematis (garis lurus terbaik) antara X (Jam Belajar) dan y (Nilai Ujian).
model.fit(X, y)

# Setelah baris ini dijalankan, model kita sudah 'pintar' dan siap untuk memprediksi.
print("Model sudah selesai dilatih!")

# 4. MEMBUAT PREDIKSI
# Kita coba gunakan model yang sudah dilatih untuk melihat hasil prediksinya pada data asli.
y_pred = model.predict(X)

# y_pred akan berisi nilai-nilai yang diprediksi oleh model untuk setiap jam belajar di data X.
print("Hasil prediksi untuk data latihan:", y_pred)


# 5. VISUALISASI HASIL
# Menggambar grafik agar kita bisa melihat seberapa baik model kita.

# plt.scatter() digunakan untuk membuat plot titik-titik (scatter plot) dari data asli.
plt.scatter(X, y, color='blue', label='Data Asli')

# plt.plot() digunakan untuk menggambar garis regresi hasil prediksi model.
# Garis ini adalah representasi dari 'pemahaman' model terhadap hubungan antara jam belajar dan nilai.
plt.plot(X, y_pred, color='red', linewidth=2, label='Garis Regresi (Prediksi Model)')

# Memberi judul pada grafik.
plt.title('Regresi Linear: Jam Belajar vs. Nilai Ujian')

# Memberi label untuk sumbu-x.
plt.xlabel('Jam Belajar')

# Memberi label untuk sumbu-y.
plt.ylabel('Nilai Ujian')

# Menampilkan legenda (keterangan warna).
plt.legend()

# Menampilkan grafik ke layar.
plt.show()


# 6. MENGGUNAKAN MODEL UNTUK MEMPREDIKSI DATA BARU
# Ini adalah tujuan utama kita: menggunakan model untuk sesuatu yang belum pernah ia lihat.

# Misalkan ada siswa baru yang belajar selama 7 jam. Berapa prediksi nilainya?
# Kita harus mengubahnya ke format array 2D seperti data latihan [[nilai]].
jam_belajar_baru = np.array([[7]])

# Menggunakan model untuk memprediksi nilai dari data baru.
prediksi_skor = model.predict(jam_belajar_baru)

# Mencetak hasil prediksi ke layar dengan format yang mudah dibaca.
# f-string (f"...") memungkinkan kita memasukkan variabel langsung ke dalam teks.
# prediksi_skor[0] mengambil elemen pertama dari hasil prediksi.
print(f"Prediksi nilai untuk siswa yang belajar selama {jam_belajar_baru[0][0]} jam adalah: {prediksi_skor[0]:.2f}")

# Angka :.2f di atas berarti kita memformat angka menjadi 2 desimal di belakang koma.
```
