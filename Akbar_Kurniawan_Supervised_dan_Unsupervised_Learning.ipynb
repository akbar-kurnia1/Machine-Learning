#Supervised Learning

```
# SUPERVISED LEARNING: REGRESI LINEAR SEDERHANA
# Tujuan: Memprediksi Nilai Ujian berdasarkan Jam Belajar

# 1. MENGIMPOR LIBRARY YANG DIBUTUHKAN
# Library adalah kumpulan kode yang sudah jadi untuk dapat digunakan ulang.

import numpy as np  # Numpy digunakan untuk operasi angka, terutama untuk membuat array (kumpulan data)
import matplotlib.pyplot as plt  # Matplotlib digunakan untuk membuat grafik atau plot visualisasi data
from sklearn.linear_model import LinearRegression # Scikit-learn (sklearn) adalah library utama untuk machine learning di Python. Mengimpor model Regresi Linear.

# 2. MENYIAPKAN DATASET
# Membuat data sederhana untuk latihan.
# Contoh data dari beberapa siswa.

# X adalah fitur (variabel independen), yaitu data yang digunakan untuk memprediksi.
# Di sini, X adalah 'Jam Belajar'.
# np.array([...]) mengubah list biasa menjadi array Numpy.
# .reshape(-1, 1) mengubah format data agar dapat dibaca oleh model scikit-learn.
X = np.array([1, 2, 3, 4, 5, 6, 8, 10]).reshape(-1, 1)

# y adalah label (variabel dependen), yaitu data yang ingin diprediksi.
# Di sini, y adalah 'Nilai Ujian'.
y = np.array([55, 60, 62, 75, 78, 85, 90, 98])

# 3. MEMBUAT DAN MELATIH MODEL
# Tahap proses 'belajar' oleh mesin.

# Membuat objek model Regresi Linear.
model = LinearRegression()

# Melatih model dengan data yang ada. Fungsi .fit() mencari pola matematis (garis lurus terbaik) antara X (Jam Belajar) dan y (Nilai Ujian).
model.fit(X, y)

# Setelah baris ini dijalankan, model telah dilatih dan siap untuk memprediksi.
print("Model sudah selesai dilatih!")

# 4. MEMBUAT PREDIKSI
# Menggunakan model yang sudah dilatih untuk melihat hasil prediksi pada data asli.
y_pred = model.predict(X)

# y_pred akan berisi nilai-nilai yang diprediksi oleh model untuk setiap jam belajar di data X.
print("Hasil prediksi untuk data latihan:", y_pred)


# 5. VISUALISASI HASIL
# Menggambar grafik untuk melihat performa model.

# plt.scatter() digunakan untuk membuat plot titik-titik (scatter plot) dari data asli.
plt.scatter(X, y, color='blue', label='Data Asli')

# plt.plot() digunakan untuk menggambar garis regresi hasil prediksi model.
# Garis ini merepresentasikan hubungan yang dipelajari model antara jam belajar dan nilai.
plt.plot(X, y_pred, color='red', linewidth=2, label='Garis Regresi (Prediksi Model)')

# Memberi judul pada grafik.
plt.title('Regresi Linear: Jam Belajar vs. Nilai Ujian')

# Memberi label untuk sumbu-x.
plt.xlabel('Jam Belajar')

# Memberi label untuk sumbu-y.
plt.ylabel('Nilai Ujian')

# Menampilkan legenda (keterangan warna).
plt.legend()

# Menampilkan grafik ke layar.
plt.show()


# 6. MENGGUNAKAN MODEL UNTUK MEMPREDIKSI DATA BARU
# Tujuan utamanya adalah menggunakan model untuk data yang belum pernah dilihat sebelumnya.

# Contoh kasus: siswa baru belajar selama 7 jam.
# Data baru harus diubah ke format array 2D seperti data latihan [[nilai]].
jam_belajar_baru = np.array([[7]])

# Menggunakan model untuk memprediksi nilai dari data baru.
prediksi_skor = model.predict(jam_belajar_baru)

# Mencetak hasil prediksi ke layar dengan format yang mudah dibaca.
# f-string (f"...") memungkinkan penyisipan variabel langsung ke dalam teks.
# prediksi_skor[0] mengambil elemen pertama dari hasil prediksi.
print(f"Prediksi nilai untuk siswa yang belajar selama {jam_belajar_baru[0][0]} jam adalah: {prediksi_skor[0]:.2f}")

# Angka :.2f di atas memformat angka menjadi 2 desimal di belakang koma.
```

#Unsupervised Learning

```
# UNSUPERVISED LEARNING: K-MEANS CLUSTERING
# Tujuan: Mengelompokkan Pelanggan berdasarkan Pendapatan dan Skor Pengeluaran

# 1. MENGIMPOR LIBRARY YANG DIBUTUHKAN
import numpy as np # Numpy untuk operasi numerik dan membuat data
import matplotlib.pyplot as plt # Matplotlib untuk membuat visualisasi atau plot
from sklearn.cluster import KMeans # Dari scikit-learn, mengimpor model K-Means

# 2. MENYIAPKAN DATASET
# Membuat data pelanggan sederhana. Setiap titik data memiliki dua nilai:
# [Pendapatan Tahunan (dalam jutaan IDR), Skor Pengeluaran (1-100)]
# Data ini tidak memiliki label 'jawaban', model harus mencari sendiri polanya.
X = np.array([
    [15, 39], [15, 81], [16, 6], [16, 77], [17, 40], [17, 76], # Grup 1 (Pendapatan rendah)
    [40, 42], [42, 55], [43, 41], [44, 59], [46, 50], [47, 45], # Grup 2 (Pendapatan menengah)
    [70, 58], [71, 95], [72, 70], [75, 90], [78, 22], [80, 88]  # Grup 3 (Pendapatan tinggi)
])

# 3. MEMBUAT DAN MENJALANKAN MODEL K-MEANS
# Tahap mengelompokkan data di atas menjadi beberapa cluster.

# Membuat objek model KMeans.
# n_clusters=3 artinya mesin akan menemukan 3 kelompok/cluster dari data.
# random_state=42 digunakan agar hasil pengelompokan selalu sama setiap kali kode dijalankan (untuk konsistensi).
# n_init='auto' adalah pengaturan teknis agar KMeans mencari hasil terbaik.
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')

# Melatih model dan langsung memprediksi (mengelompokkan) setiap titik data.
# Fungsi .fit_predict() akan melatih model dan memberikan label cluster (0, 1, atau 2) untuk setiap data.
y_kmeans = kmeans.fit_predict(X)

# y_kmeans sekarang berisi array yang menunjukkan setiap pelanggan masuk ke cluster mana.
print("Label cluster untuk setiap pelanggan:", y_kmeans)

# 4. VISUALISASI HASIL CLUSTERING
# Menggambar grafik untuk melihat hasil pengelompokan secara visual.

# Mengambil koordinat pusat dari setiap cluster yang ditemukan oleh model.
centers = kmeans.cluster_centers_
print("\nKoordinat pusat cluster:\n", centers)

# Membuat scatter plot dari data pelanggan.
# X[:, 0] mengambil semua baris, kolom pertama (Pendapatan).
# X[:, 1] mengambil semua baris, kolom kedua (Skor Pengeluaran).
# c=y_kmeans memberikan warna yang berbeda untuk setiap cluster berdasarkan hasil prediksi.
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', label='Pelanggan')

# Membuat plot untuk pusat-pusat cluster (centroids).
# Pusat cluster digambar dengan warna merah dan penanda 'X' agar menonjol.
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X', label='Pusat Cluster')

# Memberi judul dan label pada grafik agar mudah dibaca.
plt.title('Hasil Clustering Pelanggan')
plt.xlabel('Pendapatan Tahunan (jutaan IDR)')
plt.ylabel('Skor Pengeluaran (1-100)')
plt.legend()

# Menampilkan grafik ke layar.
plt.show()
```
