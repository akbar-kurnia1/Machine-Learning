```
# Import library yang dibutuhkan
# NumPy adalah library untuk operasi numerik, terutama untuk array dan matriks.
import numpy as np
# Pandas adalah library untuk manipulasi dan analisis data, khususnya untuk DataFrame.
import pandas as pd
# Scikit-learn adalah library machine learning yang populer di Python.
# datasets: Modul ini berisi beberapa dataset bawaan yang bisa digunakan untuk latihan.
from sklearn import datasets
# train_test_split: Fungsi ini digunakan untuk membagi dataset menjadi subset pelatihan dan pengujian.
from sklearn.model_selection import train_test_split
# KNeighborsClassifier: Ini adalah algoritma K-Nearest Neighbors untuk klasifikasi.
from sklearn.neighbors import KNeighborsClassifier
# accuracy_score: Fungsi ini digunakan untuk menghitung akurasi model.
from sklearn.metrics import accuracy_score
# StandardScaler: Digunakan untuk melakukan penskalaan fitur (standardisasi).
from sklearn.preprocessing import StandardScaler
# Matplotlib: Library untuk membuat visualisasi data seperti grafik dan plot.
import matplotlib.pyplot as plt
# Seaborn: Library visualisasi data berdasarkan Matplotlib, menawarkan tampilan yang lebih menarik.
import seaborn as sns

# 1. Muat Dataset
# Dataset Iris adalah dataset klasik dalam machine learning, berisi pengukuran bunga Iris.
# Tujuannya adalah mengklasifikasikan spesies Iris berdasarkan fitur-fiturnya.
iris = datasets.load_iris()

# Tampilkan deskripsi dataset
# print(iris.DESCR)

# Pisahkan fitur (X) dan target (y)
# X (fitur): Data input yang akan digunakan model untuk belajar.
#            iris.data berisi empat fitur numerik: sepal length, sepal width, petal length, petal width.
X = iris.data
# y (target): Label atau kelas yang ingin diprediksi oleh model.
#            iris.target berisi angka (0, 1, 2) yang merepresentasikan tiga spesies Iris.
y = iris.target

# Tampilkan nama fitur dan nama target
# print("Nama Fitur:", iris.feature_names)
# print("Nama Target:", iris.target_names)

# Buat DataFrame dari fitur untuk memudahkan eksplorasi dan visualisasi
df = pd.DataFrame(X, columns=iris.feature_names)
# Tambahkan kolom target ke DataFrame
df['species'] = y

# Tampilkan 5 baris pertama dari DataFrame untuk melihat strukturnya
print("5 baris pertama dari dataset:")
print(df.head())

# Tampilkan informasi dasar tentang DataFrame
print("\nInformasi DataFrame:")
df.info()

# Tampilkan statistik deskriptif dari fitur numerik
print("\nStatistik Deskriptif:")
print(df.describe())

# 2. Eksplorasi Data (Visualisasi)
# Visualisasi hubungan antar fitur dan distribusi spesies.

# Pairplot untuk melihat hubungan antara semua pasangan fitur dan distribusi setiap fitur berdasarkan spesies.
print("\nMembuat Pairplot dataset Iris...")
sns.pairplot(df, hue='species', palette='viridis')
plt.suptitle('Pairplot Dataset Iris (Fitur vs. Fitur, diwarnai berdasarkan Spesies)', y=1.02) # Tambahkan judul
plt.show() # Tampilkan plot

# Boxplot untuk melihat distribusi setiap fitur berdasarkan spesies
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i + 1) # Buat subplot 2x2
    sns.boxplot(x='species', y=feature, data=df, palette='viridis')
    plt.title(f'Distribusi {feature} berdasarkan Spesies')
plt.tight_layout() # Sesuaikan layout agar tidak tumpang tindih
plt.show()

# 3. Pra-pemrosesan Data: Penskalaan Fitur
# Penskalaan fitur penting untuk algoritma yang sensitif terhadap skala fitur (misalnya, KNN, SVM, Neural Network).
# StandardScaler menstandarisasi fitur dengan menghapus rata-rata dan menskalakan ke varian unit.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X) # Lakukan penskalaan pada data fitur

# 4. Bagi Dataset menjadi Data Pelatihan dan Data Pengujian
# Data pelatihan digunakan untuk melatih model.
# Data pengujian digunakan untuk mengevaluasi kinerja model pada data yang belum pernah dilihat.
# test_size=0.30 berarti 30% data akan digunakan sebagai data pengujian, dan 70% sebagai data pelatihan.
# random_state=42 memastikan bahwa pembagian data akan selalu sama setiap kali kode dijalankan,
# sehingga hasilnya dapat direproduksi.
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=42)

print(f"\nUkuran data pelatihan (X_train): {X_train.shape}")
print(f"Ukuran label pelatihan (y_train): {y_train.shape}")
print(f"Ukuran data pengujian (X_test): {X_test.shape}")
print(f"Ukuran label pengujian (y_test): {y_test.shape}")

# 5. Inisialisasi dan Latih Model (K-Nearest Neighbors)
# KNN adalah algoritma berbasis "jarak". Ia mengklasifikasikan titik data baru
# berdasarkan mayoritas kelas dari 'k' tetangga terdekatnya.
# n_neighbors=5 berarti model akan mempertimbangkan 5 tetangga terdekat.
knn = KNeighborsClassifier(n_neighbors=5)

# Latih model menggunakan data pelatihan
# Metode .fit() adalah di mana model "belajar" dari data.
knn.fit(X_train, y_train)

print("\nModel KNN berhasil dilatih.")

# 6. Lakukan Prediksi
# Gunakan model yang sudah dilatih untuk memprediksi kelas pada data pengujian
y_pred = knn.predict(X_test)

print("\nPrediksi pada data pengujian:")
# Tampilkan 10 prediksi pertama dan label sebenarnya untuk perbandingan
for i in range(10):
    print(f"Prediksi: {iris.target_names[y_pred[i]]}, Sebenarnya: {iris.target_names[y_test[i]]}")

# 7. Evaluasi Model
# Evaluasi seberapa baik kinerja model. Akurasi adalah metrik umum untuk klasifikasi.
# Akurasi dihitung sebagai jumlah prediksi yang benar dibagi dengan total jumlah prediksi.
accuracy = accuracy_score(y_test, y_pred)

print(f"\nMetrik Evaluasi Model:")
print(f"Akurasi Model K-Nearest Neighbors: {accuracy:.4f}") # Tampilkan akurasi dengan 4 desimal

# 8. Contoh Penggunaan Model untuk Prediksi Data Baru
# Misalkan kita punya data baru (sepal length, sepal width, petal length, petal width)
# Penting: Data baru harus diskalakan menggunakan scaler yang sama yang digunakan pada data pelatihan.
new_data = np.array([[5.1, 3.5, 1.4, 0.2]]) # Contoh data untuk Iris Setosa
# Atau contoh lain untuk Iris Versicolor/Virginica
# new_data = np.array([[6.0, 2.7, 4.2, 1.5]]) # Contoh data untuk Iris Versicolor
# new_data = np.array([[6.3, 3.3, 6.0, 2.5]]) # Contoh data untuk Iris Virginica

# Penskalaan data baru
new_data_scaled = scaler.transform(new_data)

# Prediksi menggunakan model
prediction = knn.predict(new_data_scaled)

# Dapatkan nama spesies dari prediksi
predicted_species = iris.target_names[prediction[0]]

print(f"\nPrediksi untuk data baru {new_data[0]} (setelah penskalaan):")
print(f"Spesies yang diprediksi: {predicted_species}")
```
